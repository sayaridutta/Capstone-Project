# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

import matplotlib.pyplot as plt
# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df['ConvertedComp'] = df['ConvertedComp'].fillna(0)
# Plot the distribution of the 'ConvertedComp' column

plt.hist(df['ConvertedComp'])
plt.ylabel('Frequency')
plt.xlabel('ConvertedComp')
plt.title('Distribution of ConvertedComp in the dataframe')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

import matplotlib.pyplot as plt
# print how many rows are there in the table named 'master'
QUERY = """
SELECT Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df['Age'] = df['Age'].fillna(0)

# Create the boxplot
plt.boxplot(df['Age'])
plt.ylabel('Age')
plt.title('Boxplot of Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# print how many rows are there in the table named 'master'
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df['Age'] = df['Age'].fillna(0)
df['WorkWeekHrs'] = df['WorkWeekHrs'].fillna(0)

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(df['Age'], df['WorkWeekHrs'], color='blue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.grid(True)  # Add grid lines for better readability

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# print how many rows are there in the table named 'master'
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df['Age'] = df['Age'].fillna(0)
df['WorkWeekHrs'] = df['WorkWeekHrs'].fillna(0)
df['CodeRevHrs'] = df['CodeRevHrs'].fillna(0)

# Create the bubble plot
plt.figure(figsize=(8, 6))  # Set the figure size
areas = (df['Age'] * 0.25) ** 2  # Calculate area based on Age (larger Age -> larger bubble)
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=areas, alpha=0.7, c='blue')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble size based on Age)')
plt.grid(True)  # Add grid lines for better readability

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# print rows in the table named 'DatabaseDesireNextYear'
QUERY = """
SELECT Respondent, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df['Respondent'] = df['Respondent'].fillna(0)
df['DatabaseDesireNextYear'] = df['DatabaseDesireNextYear'].fillna('Not Available')

# Sort by 'Column1' in ascending order (default)
df_sorted = df.sort_values('Respondent', ascending=False)

# Get the top 5 databases

top_5_databases = df_sorted.head(5)

# Create the pie chart
plt.pie(top_5_databases['Respondent'],
        labels=top_5_databases['DatabaseDesireNextYear'],
        autopct='%1.1f%%',
        startangle=140)
plt.axis('equal')  # Equal aspect ratio for a circular pie
plt.title('Top 5 Databases to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

df['Age'] = df['Age'].fillna(0)
df['WorkWeekHrs'] = df['WorkWeekHrs'].fillna(0)
df['CodeRevHrs'] = df['CodeRevHrs'].fillna(0)

df_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median WorkWeekHrs and CodeRevHrs
median_values = df_filtered[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create a DataFrame for plotting
df_plot = pd.DataFrame({'Hours': median_values})

# Create the stacked bar chart
df_plot.T.plot(kind='bar', stacked=True)

# Add labels and title
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.ylabel('Hours')
plt.xlabel('Hour Type')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

df['Age'] = df['Age'].fillna(0)
df['ConvertedComp'] = df['ConvertedComp'].fillna(0)

# Filter data for age group 45 to 60
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate median ConvertedComp
median_comp = filtered_df['ConvertedComp'].median()

# Group by age and calculate median ConvertedComp
median_comp_by_age = filtered_df.groupby('Age')['ConvertedComp'].median()

# Create the line chart
median_comp_by_age.plot(kind='line', figsize=(10, 6))

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp by Age (45-60)')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

# Count the occurrences of each value in the 'MainBranch' column
branch_counts = df['MainBranch'].value_counts()

# Create the horizontal bar chart
branch_counts.plot(kind='barh', figsize=(10, 6))

# Add labels and title
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branches')

plt.show()
```
